First let's read the attached data file:

```python
import pandas as pd

df0 = pd.read_csv('student-mat.csv', sep = ';')

df0.describe()
```

The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- G3  (the student's final mathematic grade)

**IVs**:
- Studytime ("studytime" column in the CSV file)
- absences  ("absences" column in the CSV file)
- health  ("health" column in the CSV file)
- Medu  ("Medu" column in the CSV file)
- Fedu  ("Fedu" column in the CSV file)


<br>
<br>

Let's extract the relevant columns:

```python

df = df0[  ['studytime', 'absences', 'health', 'Medu', 'Fedu', 'G3']  ]

df.describe()
```
