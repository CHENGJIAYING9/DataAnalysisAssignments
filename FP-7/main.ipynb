## Original vs Improved

```python
import pandas as pd

rows = []

rows.append({
    "analysis": "All students",
    "baseline_R2": base_all["R2"],
    "baseline_RMSE": float(base_all["RMSE"]),
    "improved_R2": imp_all["R2"],
    "improved_RMSE": imp_all["RMSE"],
    "delta_R2": imp_all["R2"] - base_all["R2"],
    "delta_RMSE": imp_all["RMSE"] - float(base_all["RMSE"])
})

for g in ["Low-grade", "High-grade"]:
    if g in base_groups and g in imp_groups:
        rows.append({
            "analysis": g,
            "baseline_R2": base_groups[g]["R2"],
            "baseline_RMSE": float(base_groups[g]["RMSE"]),
            "improved_R2": imp_groups[g]["R2"],
            "improved_RMSE": imp_groups[g]["RMSE"],
            "delta_R2": imp_groups[g]["R2"] - base_groups[g]["R2"],
            "delta_RMSE": imp_groups[g]["RMSE"] - float(base_groups[g]["RMSE"])
        })

summary = pd.DataFrame(rows)
summary
```

## Main reports

An initial exploration of the data shows that the distribution of absences is highly skewed, with several extreme values. These outliers can negatively affect model fitting and prediction accuracy. To mitigate this issue, we apply a 95% winsorization (clipping) to the absences variable.
